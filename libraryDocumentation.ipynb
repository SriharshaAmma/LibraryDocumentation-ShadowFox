**Introduction to Data Visualization Libraries:**

**Matplotlib**

Overview: Matplotlib is the most widely used Python library for 2D plotting. It offers low-level control over every aspect of a figure.

**Use Cases:**

* Static charts for publications and reports

* Custom visualizations for detailed control

* Base for other libraries like Seaborn

**Seaborn**

Overview: Seaborn is built on top of Matplotlib and Pandas. It provides a high-level interface for attractive and informative statistical graphics.

**Use Cases:**

* Data exploration and analysis

* Plotting complex relationships easily

* Works well with Pandas DataFrames

**Line Plot**

**Description:** Displays information as a series of data points connected by straight lines.

**Use Case:** Best for showing trends over time (e.g., stock prices, temperature changes)

**Matplotib Line Chart**

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D", "E"]
y = [20, 45, 33, 70, 25]

plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.title("Line Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

**Bar Chart**

**Description:** Uses rectangular bars to represent values of categories.

**Use Case:** Great for comparing quantities across different categories.

***Matplotib Bar Chart***

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D", "E"]
y = [20, 45, 33, 70, 25]

plt.bar(x, y, color='green')
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

 **Pie Chart**

**Description:** Circular chart divided into slices representing proportions.

**Use Case:** Useful for showing percentage distribution of a whole.

***Matplotib Pie Chart***

```python
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D", "E"]
sizes = [20, 45, 33, 70, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")
plt.axis("equal")
plt.show()
```

**Scatter Plot (Seaborn)**

**Description:** Shows the relationship between two numerical variables.

**Use Case:** Identifying correlation or clustering in data.

**Seaborn Scatter Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.title("Seaborn Scatter Plot")
plt.show()
```

**Histogram / Distribution Plot (Seaborn)**

**Description:** Shows the frequency distribution of a dataset.

**Use Case:** Understand the shape (skewness, spread) of numerical data.

**Seaborn Histogram with KDE**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.histplot(iris["petal_length"], kde=True, color="purple", bins=20)
plt.title("Histogram with KDE")
plt.show()
```

 **Box Plot (Seaborn)**

**Description:** Displays the distribution using quartiles, showing median, outliers, and range.

**Use Case:** Identify outliers and variability in the data.



**SeaBorn Box Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Box Plot")
plt.show()
```

**Heatmap (Seaborn)**

**Description:** Shows a matrix with values represented by color intensities.

**Use Case:** Best for correlation matrices, confusion matrices, or performance grids.

**SeaBorn HeatMap**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])

corr = numeric_iris.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap - Correlation Matrix")
plt.show()
```

**Line Plot (with Multiple Lines)**

This graph allows users to visually compare how:

* A linear function increases at a constant rate,

* A quadratic function increases more rapidly as x increases, and

* A cubic function increases even faster, showing exponential-like growth.



```python


x = np.linspace(0, 2, 100)  # Generates 100 values from 0 to 2

# Create a figure and a set of subplots (Axes)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# Plot three different curves on the same axes
ax.plot(x, x, label='linear')         # y = x
ax.plot(x, x**2, label='quadratic')   # y = x^2
ax.plot(x, x**3, label='cubic')       # y = x^3

# Labeling
ax.set_xlabel('x label')              # X-axis label
ax.set_ylabel('y label')              # Y-axis label
ax.set_title("Simple Plot")           # Title of the plot
ax.legend()                           # Add a legend
```
